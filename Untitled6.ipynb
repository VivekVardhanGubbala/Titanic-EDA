```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x=df['Fare'])
plt.title('Fare Distribution')
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Age vs Fare')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(),
            annot=True,
            cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(
    df[['Survived','Age','Fare','Pclass']],
    hue='Survived'
)
plt.show()
```

SUMMARY OF FINDINGS

1. Dataset contains 891 passengers.
2. Cabin column had many missing values.
3. Most passengers belonged to third class.
4. Females had significantly higher survival rates.
5. First-class passengers survived more frequently.
6. Fare contains several outliers.
7. Most passengers were aged between 20–40 years.
8. Passenger class and gender strongly influenced survival.
9. Higher fare passengers had better survival chances.

CONCLUSION:
Gender, Passenger Class, and Fare were the most important factors affecting survival on the Titanic.
